```python
# write a function named as sum_arr to find sum of elements in given array
def sum_arr(arr):
    """
    Finds the sum of all elements in an array.
    Time Complexity: O(n) - linear time
    """
    total = 0
    for num in arr:
        total += num
    return total
```

```python
#call the function
#arr = [12, 3, 4, 15,45,78,450]
#print(sum_arr)
def max_min(arr):
    """
    Finds both minimum and maximum elements in an array.
    Time Complexity: O(n) - linear time (single pass)
    Alternative: Could use min() and max() functions but that would be O(2n)
    """
    if not arr:
        return None, None

    min_val = arr[0]
    max_val = arr[0]

    for num in arr[1:]:
        if num < min_val:
            min_val = num
        if num > max_val:
            max_val = num

    return min_val, max_val
```

```python
# write a function named as Binary_Search to perform binary search in 1D-array
def Binary_Search(arr, target):
    """
    Performs binary search to find target in a SORTED array.
    Returns index if found, -1 if not found.
    Time Complexity: O(log n)
    """
    # First sort the array (binary search requires sorted data)
    sorted_arr = sorted(arr)
    left, right = 0, len(sorted_arr) - 1

    while left <= right:
        mid = (left + right) // 2

        if sorted_arr[mid] == target:
            # Find the index in the original array
            for i in range(len(arr)):
                if arr[i] == target:
                    return i
            return mid
        elif sorted_arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return -1

```
